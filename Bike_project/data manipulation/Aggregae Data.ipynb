```python
import pandas as pd

df = pd.read_excel("C:/Users/DJimenez/Desktop/visual/Bike_project/data/export_data_Jan_Jun_2021.xlsx",  skiprows=3)

def aggregate(df,name):
    df['Flow Name'] = pd.to_datetime(df['Flow Name'], format = '%Y.%m.%d %H:%M')
    #df['Flow Name'] = pd.to_datetime(df['Flow Name']).dt.date
    df['date'] = pd.to_datetime(df['Flow Name']).dt.to_period('M')
    df = df.groupby(["date"]).sum()
    df = df.T
    #Look in the data manipulation folder for out csv file
    df.to_csv("{}.csv".format(name))
    
aggregate(df,"jan_jun_2021_data")
```
